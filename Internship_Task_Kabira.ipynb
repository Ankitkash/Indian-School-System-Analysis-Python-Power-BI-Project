```python
# Total School by state
school_by_state = df['state'].value_counts().reset_index()
# Without reset_index(), the index might be like [0, 3, 5, 8]
# With reset_index(), it becomes [0, 1, 2, 3]
school_by_state.columns = ['state', 'total_schools']
school_by_state
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.bar(school_by_state['state'], school_by_state['total_schools'], color='skyblue')
plt.xticks(rotation=90)
plt.title('Total Schools by State')
plt.xlabel('State')
plt.ylabel('Number of Schools')
plt.tight_layout()
plt.show()
```

# OR

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(data=school_by_state, x='state', y='total_schools', hue = 'state', palette='viridis')
plt.xticks(rotation=90)
plt.title('Total Schools by State')
plt.xlabel('State')
plt.ylabel('Number of Schools')
plt.tight_layout()
plt.show()
```

```python
# Goverment vs Private School Count
school_dist = df['state_mgmt'].value_counts().reset_index()
school_dist
```

```python
school_dist.columns = ['Management_Type', 'School_Count']
school_dist
```

```python
# State-wise goverment & private school
type_by_state = df.groupby(['state', 'state_mgmt']).size().unstack(fill_value = 0)
type_by_state
```

```python
# School status
status_dist = df['school_status'].value_counts().reset_index()
status_dist.columns = ['school_status', 'counts']
status_dist
```

```python
# Plot
plt.figure(figsize=(8, 8))
sns.barplot(x='school_status', y='counts', hue = 'school_status', data=status_dist, palette='viridis')

plt.title('Distribution of School Statuses in Dataset')
plt.ylabel('Number of Schools')
plt.xlabel('School Status')
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```
